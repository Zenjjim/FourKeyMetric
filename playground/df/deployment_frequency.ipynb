```python
import requests
import json
import os
from dotenv import load_dotenv
import pandas as pd

load_dotenv()

def get_auth_header(token):
    print(token)
    return {"Authorization": f"Basic {token}"}

def assert_ok(http_response):
    status = http_response.status_code
    if status != 200:
        raise Exception(f"HTTP status: {status}. Body: {http_response.text}")
    return http_response.json()

def get_devops_pipeline_data(project, definition, branch, token):
    headers = get_auth_header(token)
    path = (
        f"https://dev.azure.com/{project}/_apis/build/builds?definitions={definition}&branchName={branch}&resultFilter=succeeded&statusFilter=completed&queryOrder=queueTimeDescending&api-version=6.0"
    )
    
    r = requests.get(
        path,
        headers=headers,
    )
    data = assert_ok(r)
    return data
```

```python
BRANCH = "refs/heads/main".replace("/", "%2F")
DEFINITION_ID = 82
PROJECT = "mollerdigital/carcare" 

TOKEN = os.environ["AZURE_TOKEN"]
data = get_devops_pipeline_data(PROJECT, DEFINITION_ID, BRANCH, TOKEN)
data
df = pd.DataFrame(data["value"])
```

```python
im_am_stupid = df.to_dict('records')
im_am_stupid
```

```python
release_per_day = []
import math
for item in im_am_stupid:
    try:
        index = release_per_day.index(next(filter(lambda x: x.get('date') == item.get('date'), release_per_day)))
        release_per_day[index]['count'] +=  1
    except:
        release_per_day.append({'date': item.get('date'), 'count': int(not math.isnan(item.get('id')))})
release_per_day = sorted(release_per_day, key=lambda d: d['date']) 
release_per_day = pd.DataFrame(release_per_day)
release_per_day
```

```python
release_per_day['week_start_date'] =  pd.to_datetime(release_per_day['date']) - pd.to_timedelta(7, unit='d')
release_per_day = release_per_day.groupby(['date', pd.Grouper(key='week_start_date', freq='W-MON')]).sum().reset_index()
```

```python
release_days_per_week = release_per_day.groupby('week_start_date').sum().reset_index()
release_days_per_week
```

```python
median_days_per_week = release_days_per_week['count'].median()
```

```python
releases_per_month = release_days_per_week.groupby(pd.Grouper(key='week_start_date', freq='MS')).sum().reset_index()
releases_per_month = releases_per_month.rename(columns={"week_start_date": "month_start_date"})
releases_per_month
```

```python
median_days_per_month = releases_per_month['count'].median()
median_days_per_month
```

```python
# Find median of deployment days in week.
# If more or equals then 3 then daily
# If atleast ones each week then weekly
# If atleast ones each month then monthly
# Else yearly

def deployment_frequency(frequency_week, frequency_month):
    if (frequency_week >= 3):
        return 'Daily'
    elif (frequency_week >= 1):
        return 'Weekly'
    elif (frequency_month >= 1):
        return 'Monthly'
    else:
        return 'Yearly'
deployment_frequency(median_days_per_week, median_days_per_month)
    
```
